```python
# Importando o CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
```

```python
vectorizer.fit(apenas_letras)
```

```python
matrix = vectorizer.transform(apenas_letras)
```
